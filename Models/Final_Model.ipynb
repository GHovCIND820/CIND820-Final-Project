```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Automobile_Dataset.csv")
data = df.copy(deep='False')
data.head()
```

```python
data['normalized-losses'] = pd.to_numeric(data['normalized-losses'], errors="coerce")
data['normalized-losses'] = data['normalized-losses'].fillna((data['normalized-losses'].mean()), inplace=False)
data.head()
```

```python
data['bore'] = pd.to_numeric(data['bore'], errors="coerce")
data['bore'] = data['bore'].fillna((data['bore'].mean()), inplace=False)
data.head()
```

```python
data['stroke'] = pd.to_numeric(data['stroke'], errors="coerce")
data['stroke'] = data['stroke'].fillna((data['stroke'].mean()), inplace=False)
data.head()
```

```python
data['horsepower'] = pd.to_numeric(data['horsepower'], errors="coerce")
data['horsepower'] = data['horsepower'].fillna((data['horsepower'].mean()), inplace=False)
data.head()
```

```python
data['peak-rpm'] = pd.to_numeric(data['peak-rpm'], errors="coerce")
data['peak-rpm'] = data['peak-rpm'].fillna((data['peak-rpm'].mean()), inplace=False)
data.head()
```

```python
data['price'] = pd.to_numeric(data['price'], errors="coerce")
data['price'] = data['price'].fillna((data['price'].mean()), inplace=False)
data.head()
```

```python
data.info()
```

```python
data['num-of-doors'].value_counts()
```

```python
data['num-of-doors'] = data['num-of-doors'].replace('?','four')
data['num-of-doors'].value_counts()
```

```python
string_to_number = {"num-of-doors":     {"four": 4, "two": 2},
                "num-of-cylinders": {"four": 4, "six": 6, "five": 5, "eight": 8,
                                  "two": 2, "twelve": 12, "three":3 }}

data = data.replace(string_to_number)
data.head()
```

```python
data.info()
```

```python
data_copy = data.copy()
```

Dummy Variable Function 

```python
def dummy_var(x,df):  
    temp = pd.get_dummies(df[x], prefix=x, drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
```

```python
data_copy = dummy_var('symboling',data_copy)
data_copy = dummy_var('make',data_copy)
data_copy = dummy_var('fuel-type',data_copy)
data_copy = dummy_var('aspiration',data_copy)
data_copy = dummy_var('body-style',data_copy)
data_copy = dummy_var('drive-wheels',data_copy)
data_copy = dummy_var('engine-location',data_copy)
data_copy = dummy_var('engine-type',data_copy)
data_copy = dummy_var('fuel-system',data_copy)

data_copy.head()
```

Scaling

```python
data_copy.head()
```

```python
from sklearn.preprocessing import StandardScaler,scale

sc = StandardScaler() 

columns_to_scale = ['normalized-losses','num-of-doors','wheel-base', 'length', 'width', 'height', 'curb-weight', 'num-of-cylinders', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm','city-mpg', 'highway-mpg','price',]

data_copy[columns_to_scale] = sc.fit_transform(data_copy[columns_to_scale])

data_copy.head()
```

CSV Creation

```python
Dummy_Auto = data_copy.copy()
Dummy_Auto.to_csv('Dummy_Auto.csv')
Dummy_Auto.head()
```

Model Creation

```python
from sklearn.model_selection import train_test_split

X = data_copy.loc[:, data_copy.columns != 'price']
y = data_copy.loc[:,'price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
lm = regressor.fit(X_train, y_train)
lm
```

```python
y_pred = lm.predict(X_test)

Prediction_Dataframe = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
Prediction_Dataframe
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R Squared Score: ', metrics.r2_score(y_test, y_pred))
print('Explained Variance Score: ', metrics.explained_variance_score(y_test, y_pred))
```

```python
fig = plt.figure(figsize=(11,5))
plt.scatter(y_test,y_pred)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('y_pred', fontsize=16)

#Regression Line function
f = lambda x: x

# x values of line to plot
x = np.array(y_test)

# plot fit
plt.plot(x,f(x),lw=2.5, c="orange")
```

```python
print("Training set score: {:.2f}".format(regressor.score(X_train, y_train)))
print("Test set score: {:.2f}".format(regressor.score(X_test, y_test)))
```
