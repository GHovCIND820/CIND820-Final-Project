```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Automobile_Dataset.csv")
```

```python
data_copy = df.copy(deep='False')
data_copy.head()
```

```python
data_copy['normalized-losses'] = pd.to_numeric(data_copy['normalized-losses'], errors="coerce")
data_copy['normalized-losses'] = data_copy['normalized-losses'].fillna((data_copy['normalized-losses'].mean()), inplace=False)
data_copy.head()
```

```python
data_copy['bore'] = pd.to_numeric(data_copy['bore'], errors="coerce")
data_copy['bore'] = data_copy['bore'].fillna((data_copy['bore'].mean()), inplace=False)
data_copy.head()
```

```python
data_copy['stroke'] = pd.to_numeric(data_copy['stroke'], errors="coerce")
data_copy['stroke'] = data_copy['stroke'].fillna((data_copy['stroke'].mean()), inplace=False)
data_copy.head()
```

```python
data_copy['horsepower'] = pd.to_numeric(data_copy['horsepower'], errors="coerce")
data_copy['horsepower'] = data_copy['horsepower'].fillna((data_copy['horsepower'].mean()), inplace=False)
data_copy.head()
```

```python
data_copy['peak-rpm'] = pd.to_numeric(data_copy['peak-rpm'], errors="coerce")
data_copy['peak-rpm'] = data_copy['peak-rpm'].fillna((data_copy['peak-rpm'].mean()), inplace=False)
data_copy.head()
```

```python
data_copy['price'] = pd.to_numeric(data_copy['price'], errors="coerce")
data_copy['price'] = data_copy['price'].fillna((data_copy['price'].mean()), inplace=False)
data_copy.head()
```

```python
data_copy.info()
```

```python
data_copy["fuel-type"] = data_copy["fuel-type"].astype('category')
data_copy.dtypes
```

```python
data_copy["fuel-type-code"] = data_copy["fuel-type"].cat.codes
data_copy.head()
```

```python
data_copy["aspiration"] = data_copy["aspiration"].astype('category')
data_copy.dtypes
```

```python
data_copy["aspiration-code"] = data_copy["aspiration"].cat.codes
data_copy.head()
```

```python
data_copy['num-of-doors'].value_counts()
```

```python
data_copy['num-of-doors'] = data_copy['num-of-doors'].replace('?','four')
```

```python
data_copy.info()
```

```python
data_copy['num-of-doors'].value_counts()
```

```python
string_to_number = {"num-of-doors":     {"four": 4, "two": 2},
                "num-of-cylinders": {"four": 4, "six": 6, "five": 5, "eight": 8,
                                  "two": 2, "twelve": 12, "three":3 }}

data_copy = data_copy.replace(string_to_number)
data_copy.head()
```

```python
data_copy.info()
```

```python
data_copy["body-style"] = data_copy["body-style"].astype('category')
data_copy.dtypes
```

```python
data_copy["body-style-code"] = data_copy["body-style"].cat.codes
data_copy.head()
```

```python
data_copy["engine-location"] = data_copy["engine-location"].astype('category')
data_copy.dtypes
```

```python
data_copy["engine-locatione-code"] = data_copy["engine-location"].cat.codes
data_copy.head()
```

```python
data_copy["engine-type"] = data_copy["engine-type"].astype('category')
data_copy.dtypes
```

```python
data_copy["engine-type-code"] = data_copy["engine-type"].cat.codes
data_copy.head()
```

```python
data_copy["fuel-system"] = data_copy["fuel-system"].astype('category')
data_copy.dtypes
```

```python
data_copy["fuel-system-code"] = data_copy["fuel-system"].cat.codes
data_copy.head()
```

```python
data_copy['drive-wheels'].unique()
```

```python
data_copy = pd.get_dummies(data_copy, columns=["drive-wheels"])
data_copy.head()
```

```python
data_copy.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder

oc = OrdinalEncoder()
data_copy["make_code"] = oc.fit_transform(data_copy[["make"]])
data_copy[["make", "make_code"]].head(11)
```

Normalize Dataset Using MinMaxScaler()

```python
from sklearn.preprocessing import MinMaxScaler

df_sklearn = Cleaned_DF.copy()
column = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'num-of-cylinders', 'engine-size', 'bore', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg','fuel-type-code', 'aspiration-code',	'engine-locatione-code', 'fuel-system-code', 'drive-wheels_fwd', 'drive-wheels_rwd', 'make_code']
df_sklearn[column] = MinMaxScaler().fit_transform(np.array(df_sklearn[column]))

display(df_sklearn)
```

```python
Scaled_DF = df_sklearn.copy()
Scaled_DF.to_csv('Scaled_DF.csv')
Scaled_DF.head()
```

Model Creation


```python
from sklearn.model_selection import train_test_split

A = df_sklearn.loc[:, ['symboling', 'normalized-losses',  'num-of-doors', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'num-of-cylinders', 'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg','fuel-type-code', 'aspiration-code', 'body-style-code',	'engine-locatione-code',	'engine-type-code', 'fuel-system-code', 'drive-wheels_4wd',	'drive-wheels_fwd', 'drive-wheels_rwd', 'make_code']]
B = df_sklearn.loc[:,'price']

A_train, A_test, B_train, B_test = train_test_split(A, B, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(A_train, B_train)
```

```python
Coefficient_Dataframe = pd.DataFrame(regressor.coef_, A.columns, columns=['Coefficient'])
Coefficient_Dataframe
```

```python
print(regressor.intercept_)
```

```python
B_pred = regressor.predict(A_test)
```

```python
Prediction_Frame = pd.DataFrame({'Actual': B_test, 'Predicted': B_pred})
Prediction_Frame
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(B_test, B_pred))
print('Mean Squared Error:', metrics.mean_squared_error(B_test, B_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(B_test, B_pred)))
print('R Squared Score: ', metrics.r2_score(B_test, B_pred))
print('Explained Variance Score: ', metrics.explained_variance_score(B_test, B_pred))
```

```python
fig = plt.figure(figsize=(11,5))
plt.scatter(B_test,B_pred)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('y_pred', fontsize=16)

#Regression Line function
f = lambda x: x

# x values of line to plot
x = np.array(B_test)

# plot fit
plt.plot(x,f(x),lw=2.5, c="orange")
```
